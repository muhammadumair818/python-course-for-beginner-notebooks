# üî¢ 2. Operators

Operators perform operations on values.

## Arithmetic Operators

| Operator | Meaning | Example |
|----------|---------|----------|
| `+` | Addition | `10 + 5 = 15` |
| `-` | Subtraction | `10 - 5 = 5` |
| `*` | Multiplication | `10 * 5 = 50` |
| `/` | Division | `10 / 5 = 2.0` |
| `//` | Floor Division | `10 // 3 = 3` |
| `%` | Modulus (remainder) | `10 % 3 = 1` |
| `**` | Exponent | `2 ** 3 = 8` |

```python
# Arithmetic Operators
a = 10
b = 3

print("=== Arithmetic Operators ===")
print(f"Addition: {a} + {b} = {a + b}")
print(f"Subtraction: {a} - {b} = {a - b}")
print(f"Multiplication: {a} * {b} = {a * b}")
print(f"Division: {a} / {b} = {a / b}")
print(f"Floor Division: {a} // {b} = {a // b}")
print(f"Modulus: {a} % {b} = {a % b}")
print(f"Exponent: {a} ** {b} = {a ** b}")
```

## Comparison Operators

These return `True` or `False`.

| Operator | Meaning | Example |
|----------|---------|----------|
| `==` | Equal to | `5 == 5` |
| `!=` | Not equal to | `5 != 3` |
| `>` | Greater than | `10 > 5` |
| `<` | Less than | `3 < 5` |
| `>=` | Greater than or equal | `5 >= 5` |
| `<=` | Less than or equal | `3 <= 5` |

```python
# Comparison Operators
x = 5
y = 10

print("=== Comparison Operators ===")
print(f"{x} == {y}: {x == y}")
print(f"{x} != {y}: {x != y}")
print(f"{x} < {y}: {x < y}")
print(f"{x} > {y}: {x > y}")
print(f"{x} <= {y}: {x <= y}")
print(f"{x} >= {y}: {x >= y}")
```

## Logical Operators

| Operator | Meaning |
|----------|----------|
| `and` | Both conditions must be True |
| `or` | At least one condition must be True |
| `not` | Reverses the condition |

```python
# Logical Operators
age = 25
has_license = True
is_student = False

print("=== Logical Operators ===")
print(f"age > 18 and has_license: {age > 18 and has_license}")
print(f"is_student or has_license: {is_student or has_license}")
print(f"not is_student: {not is_student}")
print(f"age > 18 and is_student: {age > 18 and is_student}")
```

# üîÅ 3. Control Flow

Control flow lets your program make decisions and repeat actions.

## If, Elif, Else Statements

```python
# If-Elif-Else Example
score = 85

print(f"Your score: {score}")

if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
elif score >= 70:
    grade = "C"
elif score >= 60:
    grade = "D"
else:
    grade = "F"

print(f"Your grade: {grade}")

# Another example
age = 17
if age >= 18:
    print("You are an adult")
else:
    print("You are a minor")
```

## For Loops

Repeat code a specific number of times or for each item.

```python
# For Loop Examples
print("=== Loop with range() ===")
for i in range(1, 6):
    print(f"Number: {i}")

print("\n=== Loop through list ===")
fruits = ["apple", "banana", "orange"]
for fruit in fruits:
    print(f"I like {fruit}")

print("\n=== Loop with index ===")
for index, fruit in enumerate(fruits):
    print(f"{index}: {fruit}")
```

## While Loops

```python
# While Loop Example
print("=== While Loop ===")
count = 1
while count <= 5:
    print(f"Count: {count}")
    count += 1
```

## Break and Continue

```python
# Break example
print("=== Break ===")
for i in range(1, 11):
    if i == 5:
        print("Found 5! Breaking out.")
        break
    print(i)

# Continue example
print("\n=== Continue ===")
for i in range(1, 6):
    if i == 3:
        print("Skipping 3")
        continue
    print(f"Number: {i}")

# PRACTICE TASK:
# Write a loop that prints 1-10
# Skip even numbers using continue
# YOUR CODE HERE
```

# üì¶ 4. Data Structures

Data structures organize multiple values together.

## Lists

Ordered, changeable collection.

```python
# Lists
numbers = [1, 2, 3, 4, 5]
print(f"List: {numbers}")

# Access elements
print(f"First: {numbers[0]}")
print(f"Last: {numbers[-1]}")

# Slicing
print(f"Middle: {numbers[1:4]}")

# Modifying
numbers.append(6)
print(f"After append: {numbers}")

numbers.remove(3)
print(f"After remove: {numbers}")

# Methods
print(f"Length: {len(numbers)}")
print(f"Sum: {sum([1, 2, 3, 4, 5])}")
print(f"Max: {max([1, 2, 3, 4, 5])}")
```

## Tuples

Ordered, unchangeable collection.

```python
# Tuples
colors = ("red", "green", "blue")
print(f"Tuple: {colors}")

# Access
print(f"First: {colors[0]}")

# Tuples cannot be modified
# colors[0] = "yellow"  # Error!

# Unpacking
first, second, third = colors
print(f"Unpacked: {first}, {second}, {third}")
```

## Dictionaries

Key-value pairs.

```python
# Dictionaries
student = {
    "name": "Umair",
    "age": 25,
    "major": "Computer Science",
    "gpa": 3.8
}

print(f"Student: {student}")

# Access
print(f"Name: {student['name']}")
print(f"Age: {student['age']}")

# Modify
student['age'] = 26
student['city'] = "Karachi"
print(f"Updated: {student}")

# Methods
print(f"Keys: {list(student.keys())}")
print(f"Values: {list(student.values())}")
```

## Sets

# üß© 5. Functions

Functions are reusable blocks of code.

## Defining Functions

```python
# Basic function
def greet(name):
    """Function that greets someone"""
    return f"Hello, {name}!"

print(greet("Umair"))
print(greet("Alice"))

# Function with multiple parameters
def add(a, b):
    return a + b

print(f"\n5 + 3 = {add(5, 3)}")

# Function with default arguments
def power(base, exponent=2):
    return base ** exponent

print(f"2^2 = {power(2)}")
print(f"2^3 = {power(2, 3)}")
```

## Lambda Functions

Small, anonymous functions for simple operations.

```python
# Lambda function
square = lambda x: x ** 2
print(f"Square of 5: {square(5)}")

# Using lambda with map
numbers = [1, 2, 3, 4, 5]
squared = list(map(lambda x: x ** 2, numbers))
print(f"Original: {numbers}")
print(f"Squared: {squared}")

# Using lambda with filter
evens = list(filter(lambda x: x % 2 == 0, numbers))
print(f"Even numbers: {evens}")

# PRACTICE TASK:
# Create a function to convert Celsius to Fahrenheit
# Formula: F = (C * 9/5) + 32
# Test with 0, 25, 100
# YOUR CODE HERE
```

# üß† 7. Object-Oriented Programming (OOP)

Organize code into objects with properties and behaviors.

## Classes and Objects

```python
# Basic class
class Student:
    """A class representing a student"""
    
    def __init__(self, name, age, gpa):
        self.name = name
        self.age = age
        self.gpa = gpa
    
    def display_info(self):
        return f"{self.name} (Age: {self.age}, GPA: {self.gpa})"
    
    def is_honor_student(self):
        return self.gpa >= 3.5

# Create objects
student1 = Student("Umair", 25, 3.8)
student2 = Student("Alice", 24, 3.2)

print(f"Student 1: {student1.display_info()}")
print(f"Honor Student: {student1.is_honor_student()}")

print(f"\nStudent 2: {student2.display_info()}")
print(f"Honor Student: {student2.is_honor_student()}")
```

## Inheritance

```python
# Parent class
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    
    def greet(self):
        return f"Hello, I am {self.name}"

# Child class
class Employee(Person):
    def __init__(self, name, age, employee_id, salary):
        super().__init__(name, age)
        self.employee_id = employee_id
        self.salary = salary
    
    def get_details(self):
        return f"{self.greet()}. ID: {self.employee_id}, Salary: ${self.salary}"

# Create object
employee = Employee("Umair", 25, "E001", 50000)
print(employee.get_details())

# PRACTICE TASK:
# Create a BankAccount class with:
# - account_holder, balance
# - deposit() method
# - withdraw() method
# Test it
# YOUR CODE HERE
```

## Project 1: Number Guessing Game

```python
import random

def guessing_game():
    """Simple number guessing game"""
    secret = random.randint(1, 100)
    attempts = 0
    guessed = False
    
    print("\n" + "="*50)
    print("NUMBER GUESSING GAME")
    print("I am thinking of a number between 1-100")
    print("="*50)
    
    while not guessed:
        try:
            guess = int(input("Enter your guess: "))
            attempts += 1
            
            if guess < 1 or guess > 100:
                print("Enter number between 1-100!")
                continue
            
            if guess < secret:
                print(f"Too low! Try again. (Attempt {attempts})")
            elif guess > secret:
                print(f"Too high! Try again. (Attempt {attempts})")
            else:
                print(f"\nCorrect! The number was {secret}!")
                print(f"You guessed it in {attempts} attempts!")
                guessed = True
        
        except ValueError:
            print("Enter a valid number!")

print("\nNumber Guessing Game function created!")
print("Call guessing_game() to play (requires interactive input)")
```

## Project 3: Student Management

```python
import json

class StudentManager:
    def __init__(self):
        self.students = {}
    
    def add_student(self, roll_no, name, marks):
        self.students[roll_no] = {'name': name, 'marks': marks}
        print(f"Student {name} added!")
    
    def view_all(self):
        if not self.students:
            print("No students found!")
            return
        
        print("\n" + "="*50)
        print("STUDENT RECORDS")
        print("="*50)
        print(f"{'Roll':<10} {'Name':<20} {'Marks':<10}")
        print("-"*50)
        
        for roll_no, data in self.students.items():
            print(f"{roll_no:<10} {data['name']:<20} {data['marks']:<10}")
        print("="*50)
    
    def get_student(self, roll_no):
        if roll_no in self.students:
            s = self.students[roll_no]
            print(f"\nRoll: {roll_no}")
            print(f"Name: {s['name']}")
            print(f"Marks: {s['marks']}")
        else:
            print(f"Student {roll_no} not found!")

# Demo
print("=== STUDENT MANAGER DEMO ===")
manager = StudentManager()
manager.add_student('001', 'Umair', 95)
manager.add_student('002', 'Alice', 92)
manager.add_student('003', 'Bob', 87)
manager.view_all()
manager.get_student('001')
```
